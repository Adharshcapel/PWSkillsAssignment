Q5  Why and when we should use a while loop give a detailed explanation?

    - A while loop is a control flow statement that allows you to execute a block of code repeatedly as long as a specified condition is true. While loops are useful when you need to repeat a task multiple times, and the number of iterations is not known in advance.
    
- Iterating Until a Condition is Met: While loops are often used to iterate over a sequence of values until a certain condition is met. For example, you might use a while loop to search through a list until you find a specific element or until a certain criteria is satisfied.

- Repeating Until a User Input: While loops are useful for tasks that require user input validation. For example, you might prompt the user to enter a valid input, and keep prompting until they provide a valid response.

- Processing Items from an Input Stream: While loops are commonly used for processing items from an input stream, such as reading lines from a file or processing data from a network socket, until there is no more data to process.

- Implementing a Countdown or Timer: While loops can be used to implement countdowns, timers, or other time-based operations. You can repeatedly update a timer variable until a certain condition is met, such as reaching zero.

- Performing Task Until a Resource is Exhausted: While loops are useful for performing a task until a resource is exhausted, such as processing items from a queue until the queue is empty.

```python
# Q6 Use nested while loop to print three different pattern? 
    
    ## Right angle triangle
    
rows = 5
i = 0

print('Right angle triangle')  
while i < rows:
    j = 0
    while j <= i:
        print("*", end="")
        j += 1
    print()
    i += 1
  
```

```python
#  Inverted Right Triangle
rows = 5
i = 0
print('Inverted Right Triangle')
while i < rows:
    j = rows
    while j > i:
        print("*", end="")
        j -= 1
    print()
    i += 1
```

```python
## Pyramid

rows = 3
i = 0
print("Pyramid")
while i < rows:
    j = rows - i - 1
    while j > 0:
        print(" ", end="")
        j -= 1
    k = 0
    while k <= i:
        print("*", end="")
        k += 1
    l = i - 1
    while l >= 0:
        print("*", end="")
        l -= 1
    print()
    i += 1
```
